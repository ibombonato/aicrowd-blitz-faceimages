## Generating Prediction File

In this starter kit we will directly be submitting the blurred images. In case you would like to see a submission via baseline model.

```python
%cd /content/
!mkdir original
!cp -r /content/data/test/original/restored_imgs/* /content/original/
```

```python
from PIL import Image
import os, sys
from tqdm import tqdm

path = "/content/original/"
dirs = os.listdir(path)

def resize():
    for item in tqdm(dirs):
        if os.path.isfile(path+item):
            im = Image.open(path+item)
            f, e = os.path.splitext(path+item)
            imResize = im.resize((512,512), Image.ANTIALIAS)
            imResize.save(f + '.jpg', 'JPEG', quality=100)

resize()
```
